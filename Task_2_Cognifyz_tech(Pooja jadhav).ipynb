```python
import matplotlib.pyplot as plt
gender_counts = df['gender'].value_counts()
plt.bar(gender_counts.index, gender_counts.values, color='pink')

plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Gender Distribution')

plt.show()
```
